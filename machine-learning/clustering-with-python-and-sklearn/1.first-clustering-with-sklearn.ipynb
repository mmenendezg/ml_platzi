# First clustering with Scikit-Learn

## Setup

```python
from sklearn.datasets import make_blobs
import pandas as pd
import polars as pl
import matplotlib.pyplot as plt
```

## Data Processing

```python
x, y = make_blobs(n_samples=100, centers=6, n_features=2, cluster_std=[0.3, 2, 1.5, 1, 1.2, 0.8], random_state=1992)
```

```python
df_blobs = pd.DataFrame({"x1": x[:, 0], "x2": x[:, 1], "y": y})
df_blobs.head()
```

```python
def plot_clusters(x: pd.DataFrame, y: pd.Series, ax):
    y_uniques = y.unique()

    for val in y_uniques:
        x[y == val].plot(
            title=f"{len(y_uniques)} Clusters",
            kind="scatter",
            x="x1",
            y="x2",
            marker=f"${val}$",
            ax=ax
        )
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
x, y = df_blobs[["x1", "x2"]], df_blobs["y"]
plot_clusters(x, y, ax)
```
