# CSV files

```python
import pandas as pd 
import numpy as np 
```

```python
pd.__version__
```

```python
file_name = ''
dir_data = 'data/{}'.format(file_name)
```

```python
dict_data = {
    'edad' : [10, 9, 13, 14, 12, 11, 12],
    'cm' : [115, 110, 130, 155, 125, 120, 125],
    'pais' : ['co', 'mx', 'co', 'mx', 'mx', 'ch', 'ch'],
    'genero' : ['M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'Q1' : [5, 10, 8, np.nan, 7, 8, 3],
    'Q2' : [7, 9, 9, 8, 8, 8, 9]
}
dict_data
```

```python
df = pd.DataFrame(dict_data)
df 
```

```python
file_name = 'test.csv'
dir_data = 'data/{}'.format(file_name)
df.to_csv(dir_data, sep='|', index=False)
```

```python
df_read = pd.read_csv('data/test.csv', sep='|')
df_read
```

# Excel Files

```python
def file_saved(file_name):
    dir_data = 'data/{}'.format(file_name)
    return dir_data

df.to_excel(file_saved('test.xlsx'))
```

## Files JSON

```python
df.to_json(file_saved('test.json'))
```

```python
df_json = pd.read_json(file_saved('test.json'))
df_json
```

## Files pickle

```python
df.to_pickle(file_saved('test.pkl'))
df_pkl = pd.read_pickle(file_saved('test.pkl'))
df_pkl
```
