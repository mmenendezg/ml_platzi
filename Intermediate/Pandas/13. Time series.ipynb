# Time series

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/covid_19_data.csv')
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
df.sample(10)
```

```python
list(df)
```

```python
df = df[[
    'ObservationDate',
    'Country/Region',
    'Confirmed',
    'Deaths',
    'Recovered']]
df
```

```python
df_time = df.groupby('ObservationDate').sum()
df_time
```

```python
df1 = df_time['Confirmed'].iloc[10:15]
df1
```

```python
df2 = df_time['Deaths'].iloc[12:17]
df2
```

```python
df1 - df2
```

```python
df_time.diff().mean()
```

```python
df_diff = df_time.diff()
```

```python
df_time[['Confirmed', 'Deaths', 'Recovered']].head(1).to_dict()
```

```python
df_diff = df_diff.fillna({'Confirmed': 555.0,
 'Deaths': 17.0,
 'Recovered': 28.0})
```

```python
df_diff.cumsum()
```

```python
df_diff.resample('M').sum()
```

```python
df_diff.resample('M').count()
```

```python
df_cum = df_time.resample('12h').sum(min_count=1)
```

```python
df_cum.interpolate(inplace=True)
```

```python
df_cum['surv_rate'] = round(1 - df_cum['Deaths']/df_cum['Confirmed'], 4)
```

```python
df_cum = df_cum.reset_index()
```

```python
sr = df_cum.groupby(pd.Grouper(key='ObservationDate', freq='1D'))['surv_rate'].mean()
sr.plot()
```

```python
sr.plot()
sr.rolling(window=7).mean().plot()
```

## Ploting Tools in Pandas

```python
df.head()
```

```python
df.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False)
```

```python
list(df)
```

```python
df_time = df.groupby(['Country/Region', pd.Grouper(key='ObservationDate', freq='1D')]).sum()
df_time
```

```python
df_china = df_time.loc['Mainland China', :]
df_china
```

```python
import matplotlib.pyplot as plt
```

```python
df_china.plot(
    figsize=(10,7), 
    title='covid-19', 
    legend=False,
    style = ['r.', 'g--', 'b:*'],
)
plt.xlabel('Date')
plt.ylabel('People')
plt.show()
```

```python
df_monthly = df_china.resample('M').max()
df_monthly
```

```python
df_monthly.plot(
    figsize=(10,7),
    kind='bar', 
    stacked=True
)
```

```python
df_monthly['Traitements'] = df_monthly['Confirmed'] - df_monthly['Deaths'] - df_monthly['Recovered']
df_monthly
```

```python
df_monthly[['Deaths', 'Recovered', 'Traitements']].T.plot(figsize=(10,7), kind='pie', subplots=True)
```

```python
df_china['rate'] = 1 - df_china['Deaths'] / df_china['Confirmed']
df_china
```

```python
df_china['rate'].plot(figsize=(10,7), kind = 'kde')
```
