# Conceptos estadísticos analíticos

```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy import stats
from scipy.stats import norm, t, chi2, f
import seaborn as sns

%matplotlib inline
```

```python
x1 = 3
x2 = 10

mu_1 = 4
mu_2 = 7
sigma = 1

p_sample_1 = norm.pdf(x1, mu_1, sigma) * norm.pdf(x1, mu_1, sigma)
p_sample_2 = norm.pdf(x1, mu_2, sigma) * norm.pdf(x1, mu_2, sigma)

print(p_sample_1)
print(p_sample_2)
```

```python
muestra_10 = norm.rvs(5, sigma, size=10)
muestra_10
```

```python
data_1 = norm.rvs(mu_1, sigma, size=10000)
data_2 = norm.rvs(mu_2, sigma, size=10000)
```

```python
axis = sns.distplot(data_1, bins=50, color='blue')
axis.set(xlabel='Distribución normal mu_1', ylabel='Frequency')

axis = sns.distplot(data_2, bins=50, color='red')
axis.set(xlabel='Distribución normal mu_2', ylabel='Frequency')
```

## Distribuciones muestreales

```python
data_1 = t.rvs(100, size=1000000)
data_2 = t.rvs(0, size=1000000)

axis = sns.distplot(data_1, bins=500, kde=False, color='red')
axis = sns.distplot(data_2, bins=500, kde=False, color='blue')
```

```python
data_3 = f.rvs(5, 25, size=1000000)
data_4 = f.rvs(15, 25, size=1000000)

axis = sns.distplot(data_3, bins=500, kde=False, color='red')
axis = sns.distplot(data_4, bins=500, kde=False, color='blue')
```
