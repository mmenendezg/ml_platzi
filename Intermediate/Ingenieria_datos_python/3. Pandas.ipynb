## Data Wrangling

```python
# Add newspaper_uid to the Dataframe

radio_canada['newspaper_uid'] = 'radio-canada'
radio_canada
```

```python
# Get the host of the article link

from urllib.parse import urlparse

radio_canada['host'] = radio_canada['url'].apply(lambda url: urlparse(url).netloc)
radio_canada
```

## Missing Data

```python
missing_titles_mask = radio_canada['title'].isna()
missing_titles_mask
```

```python
missing_titles = (radio_canada[missing_titles_mask]['url']
                    .str.extract(r'(?P<missing_titles>[^/]+)$')
                    .applymap(lambda title: title.split('-'))
                    .applymap(lambda title_word_list: ' '.join(title_word_list))
                )

missing_titles
```

```python
radio_canada.loc[missing_titles_mask, 'title'] = missing_titles.loc[:, 'missing_titles']
radio_canada
```

## Additional Cleanup

```python
# Add uid to the rows

import hashlib

uids = (radio_canada
            .apply(lambda row: hashlib.md5(bytes(row['url'].encode())), axis=1)
            .apply(lambda hash_object: hash_object.hexdigest())
        )
radio_canada['uid'] = uids

radio_canada.set_index('uid', inplace=True)
radio_canada
```

```python
stripped_body = (radio_canada
                    .apply(lambda row: row['body'], axis=1)
                    .apply(lambda body: list(body))
                    .apply(lambda letters: list(map(lambda letter: letter.replace('“', ''), letters)))
                    .apply(lambda letters: list(map(lambda letter: letter.replace('"', ''), letters)))
                    .apply(lambda letters: ''.join(letters))
                )
stripped_body.head(20)
```

## Data enrichment

```python
# Tokenize title and body

import nltk
from nltk.corpus import stopwords
```

```python
def tokenize_column(df, column_name):
    
    return (df
                .dropna()
                .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
                .apply(lambda tokens: list(filter(lambda token: token.isalpha(), tokens)))
                .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
                .apply(lambda word_list: list(filter(lambda word: word not in fr_stop_words, word_list)))
                .apply(lambda valid_word_list: len(valid_word_list))
    )
```

```python
fr_stop_words = set(stopwords.words('french'))
en_stop_words = set(stopwords.words('english'))

radio_canada['n_tokens_title'] = tokenize_column(radio_canada, 'title')
radio_canada
```
