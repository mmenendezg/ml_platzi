```python
import scrapy

class DiarioES(scrapy.Spider):
    name = 'diario_es'
    allowed_domains = ['https://diarioelsalvador.com/']
    customed_settings = {
        'FEED_FORMAT' : 'json',
        'FEED_URI' : 'scrap_diario_es.json',
        'DEPTH_LIMIT' : 2
    }
    start_urls = [
        'https://diarioelsalvador.com/seccion/depais/',
        'https://diarioelsalvador.com/seccion/depalabra/',
        'https://diarioelsalvador.com/seccion/dedinero/',
        'https://diarioelsalvador.com/seccion/decomercio/',
        'https://diarioelsalvador.com/seccion/deplaneta/',
        'https://diarioelsalvador.com/seccion/deinnovacion/',
        'https://diarioelsalvador.com/seccion/decultura/',
        'https://diarioelsalvador.com/seccion/devida/',
        'https://diarioelsalvador.com/seccion/dediversion/',
        'https://diarioelsalvador.com/seccion/dedeportes/'
    ]
    
    def parse(self, response):
        
        featured_article = response.xpath('//div[@class="jeg_thumb"]/a/@href').get()
        if featured_article is not None:
            yield response.follow(featured_article, callback=self.parse_article)
        
        articles = response.xpath('//article[@class="jeg_posts"]//h3[@class="jeg_post_title"]//a/@href').getall()
        for article in articles:
            yield response.follow(article, callback=self.parse_article)
        
        next_page = response.xpath('//a[@class="page_nav next"]/@href')
        if next_page is not None:
            yield response.follow(next_page, callback=self.parse)
        
    
    def parse_article(self, response):
        title = response.xpath('//h1[@class="jeg_post_title"]/text()').get()
        body = response.xpath('div[@class="content-inner"]/text()').getall()
        yield {
            'url' : response.url,
            'title' : title,
            'body' : body
        }
```

```python
from scrapy.crawler import CrawlerProcess

process = CrawlerProcess()
process.crawl(DiarioES)
process.start()
```
