# Descenso del Gradiente

```python
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt
```

## Gráfica en 3D de función de coste

```python
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})

def f(x,y):
    function = x**2 + y**2
    return function

x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

surf = ax.plot_surface(X, Y, Z, cmap=cm.cool)

fig.colorbar(surf)
```

## Descenso del gradiente

```python
level_map = np.linspace(np.min(Z), np.max(Z), 100)
plt.contour(X, Y, Z, levels=level_map, cmap=cm.cool)
plt.colorbar()
plt.title('Descenso del Gradiente')

h = 0.01
learning_rate = 0.01

def derivate(cp, p):
    return (f(cp[0], cp[1]) - f(p[0], p[1])) / h

def gradient(p):
    grad = np.zeros(2)
    for idx, val in enumerate(p):
        cp = np.copy(p)
        cp[idx] = cp[idx] + h
        
        dp = derivate(cp, p)
        grad[idx] = dp

    return grad

for i in range(1000):
    p = p - learning_rate*gradient(p)

    if i%10 == 0 :
        plt.plot(p[0], p[1], 'o', c='r')
plt.plot(p[0], p[1], 'o', c='b')
```
