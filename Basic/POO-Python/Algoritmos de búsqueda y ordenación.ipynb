# Busqueda lineal $O(n)$





```python
import random

def busqueda_lineal(lista, objetivo):
    match = False

    for elemento in lista: # O(n)
        if elemento == objetivo:
            match = True
            break
    
    return match

if __name__ == '__main__':
    list_size = int(input('What is the size of the list?'))
    target = int(input('What is the target to search?'))

    lista = [random.randint (0, 100) for i in range(list_size)]

    print(lista)
    encontrado = busqueda_lineal(lista, target)
    print(f'The element {target} {"is" if encontrado else "is not"} in the list')
```

# Busqueda Binaria 

* El problema asume que los datos están ordenandos
* El problema se divide en 2 en cada iteración 

```python
import random 

def busqueda_binaria(lista, comienzo, final, objetivo):
    print(f'Buscando {objetivo} entre {lista[comienzo]} y {lista[final - 1]}')
    if comienzo > final:
        return False
    
    medio = (comienzo + final) // 2

    if lista[medio] == objetivo:
        return True
    elif lista[medio] < objetivo:
        return busqueda_binaria(lista, medio + 1, final, objetivo)
    else:
        return busqueda_binaria(lista, comienzo, medio - 1, objetivo)

if __name__ == '__main__':
    tamano_lista = int(input('What is the size of the list?'))
    objetivo = int(input('What is the target to search?'))

    lista = sorted([random.randint (0, 100) for i in range(tamano_lista)])

    encontrado = busqueda_binaria(lista, 0, (len(lista) -1 ), objetivo)

    print(lista)
    print(f'The element {objetivo} {"is" if encontrado else "is not"} in the list')
```

# Comparación de busqueda lineal y busqueda binaria

Evaluemos cual es la eficiencia de algoritmo vs el otro. 

```python
import random 

def busqueda_lineal(lista, objetivo):

    iteraciones = 0
    match = False

    for elemento in lista: # O(n)

        iteraciones += 1

        if elemento == objetivo:
            match = True
            break
    
    return match, iteraciones

def busqueda_binaria(lista, comienzo, final, objetivo, iteraciones=0):
    
    iteraciones += 1

    if comienzo > final:
        return False, iteraciones
    
    medio = (comienzo + final) // 2

    if lista[medio] == objetivo:
        return True, iteraciones
    elif lista[medio] < objetivo:
        return busqueda_binaria(lista, medio + 1, final, objetivo, iteraciones)
    else:
        return busqueda_binaria(lista, comienzo, medio - 1, objetivo, iteraciones)

if __name__ == '__main__':
    tamano_lista = int(input('Define el tamaño de la lista: '))
    objetivo = int(input('Define el objetivo a buscar: '))

    print(tamano_lista)

    lista = [random.randint (0, 100) for i in range(tamano_lista)]
    lista_ordenada = sorted(lista)
    
    resultado_lineal = busqueda_lineal(lista, objetivo)

    resultado_binario = busqueda_binaria(lista_ordenada, 0, (len(lista_ordenada) - 1), objetivo)

    print(lista_ordenada)
    print(f'El elemento {objetivo} {"está" if resultado_lineal[0] else "no está"} en la lista.')
    print(f'Iteraciones con el algoritmo de busqueda lineal: {resultado_lineal[1]}')
    print(f'Iteraciones con el algoritmo de busqueda binaria: {resultado_binario[1]}')
```

# Bubble Sort ---> $O(n^2)$

> El ordernamiento de burbuja es un algoritmmo que recorre repetidamente una lista que necesita ordernarse. Compara elementos adyacentes y los intercambia si están en el orden incorrect. Este procedimiento se repite hasta que no se requieren más intercambios, los que indica que la lista se encuentra ordenada. 

```python
import random 

def ordernamiento_burbuja(lista):
    n = len(lista)

    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]: # O(n) * O(n) = O(n^2)
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    
    return lista

if __name__ == '__main__':
    tamano_lista = int(input('Define el tamaño de la lista: '))
    lista = [random. randint(0, 100) for i in range(tamano_lista)]                          
    lista_ordenada = ordernamiento_burbuja(lista)
    print(lista_ordenada)
```

# Insertion Sort

>Una lista es dividida entre una sublista ordenada y otra sublista desordenada.
Al principio, la sublista ordenada contiene un solo elemento, por lo que por
definición se encuentra ordenada.
>
>A continuación se evalua el primer elemento dentro la sublista desordenada para
que podamos insertarlo en el lugar correcto dentro de la lista ordenada.
>
>La inserción se realiza al mover todos los elementos mayores al elemento que
se está evauluando un lugar a la derecha.
>
>Continua el proceso hasta que la sublista desordenada quede vacia y, por lo
tanto, la lista se encontrará ordenada.




```python
import random

def insertion_sort(lista):
    
    for indice in range(1, len(lista)):
        valor_actual = lista[indice]
        posicion_actual = indice

        while posicion_actual > 0 and lista[posicion_actual - 1] > valor_actual:
            lista[posicion_actual] = lista[posicion_actual - 1]
            posicion_actual -= 1
        
        lista[posicion_actual] = valor_actual
    return lista

if __name__ == '__main__':
    tamano_lista = int(input('Define el tamaño de la lista: '))
    lista = [random. randint(0, 100) for i in range(tamano_lista)]                          
    lista_ordenada = insertion_sort(lista)
    print(lista_ordenada)
```

# Merge Sort

>El ordenamiento por mezcla es un algoritmo de divide y conquista. Primero divide una lista en partes iguales hasta que quedan sublistas de 1 o 0 elementos. Luego los recombina en forma ordenada.

```python
import random

def merge_sort(lista):

    if len(lista) > 1:

        medio = len(lista) // 2
        izquierda = lista[:medio]
        derecha = lista[medio:]
        print(izquierda, '*' * 5, derecha)

        # Llamada recursiva de cada mitad
        merge_sort(izquierda)
        merge_sort(derecha)

        # Iteradores para recorrer las dos sublistas
        i = 0
        j = 0
        k = 0

        while i < len(izquierda) and j < len(derecha):

            if izquierda[i] < derecha[j]:
                lista[k] = izquierda[i]
                i += 1
            else:
                lista[k] = derecha[j]
                j += 1
            
            k += 1
        
        while i < len(izquierda):
            lista[k] = izquierda[i]
            i += 1
            k += 1
        
        while j < len(derecha):
            lista[k] = derecha[j]
            j += 1
            k += 1
        
        print(f'izquierda {izquierda}, derecha {derecha}')
        print(lista)
        print('<', '-' * 50, '>')

    return lista

if __name__ == '__main__':
    tamano_lista = int(input('Define el tamaño de la lista: '))
    lista = [random. randint(0, 100) for i in range(tamano_lista)]
    print(lista)
    print('-' * 20)
    lista_ordenada = merge_sort(lista)
    print(lista_ordenada)
```
