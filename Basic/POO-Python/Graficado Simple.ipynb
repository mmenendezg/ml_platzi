# Graficado simple

```python
from bokeh.plotting import figure, output_notebook, show

output_notebook()

if __name__ == '__main__':
    fig = figure()

    total_vals = int(input('Define cuantos valores se van a graficar: '))
    x_vals = list(range(total_vals))
    
    y_vals = []
    for x in x_vals:
        val = int(input(f'El valor y para x={x}: '))
        y_vals.append(val)
    
    fig.line(x_vals, y_vals, line_width=2)
    show(fig)
```
