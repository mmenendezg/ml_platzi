## Instancias

Una instancia es un objeto creado a partir de una clase (molde).

```
class Coordenada:

    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def distancia(self, otra_coordenada):
        x_diff = (self.x - otra_coordenada.x) ** 2
        y_diff = (self.y - otra_coordenada.y) ** 2

        return (x_diff + y_diff) ** 0.5

if __name__ == '__main__':
    coord_1 = Coordenada(3, 30)
    coord_2 = Coordenada(4, 8)

    print(coord_1.distancia(coord_2))
```

Ahora evaluamos si coord_1 es una instancia de Coordenada.

```
print(isinstance(coord_1, Coordenada))
print(isinstance(3, Coordenada))
```

## Abstracción

- Enfocarnos en la información relevante
- Separar la información central de los detalles secundarios
- Podemos utilizar variables y métodos (privados o públicos)


```
class Lavadora:

    def __init__(self):
        pass
    
    def lavar(self, temperatura='caliente'):
        self._llenar_tanque_de_agua(temperatura)
        self._add_detergent()
        self._wash()
        self._centrifugar()

    def _llenar_tanque_de_agua(self, temperatura):
        print(f'Llenando el tanque con agua {temperatura}')
    
    def _add_detergent(self):
        print('Adding detergent')

    def _wash(self):
        print('Washing the clothes')
    
    def _centrifugar(self):
        print('Centrifugando la ropa')
    
if __name__ == '__main__':
    lavadora = Lavadora()
    lavadora.lavar()
```

## Encapsulación, getters and setters 
### Encapsulación

- Permite agrupar datos y su comportamiento y controla el acceso a dichos datos previniendo modificaciones no autorizadas


```
class Coordinate:

    def __init__(self, x=0, y=0):
        self._x = x
        self._y = y
    
    @property
    def x(self):
        return self._x
    
    @x.setter
    def set_x(self, coord_x):
        # Evaluate if the coord_x is an int
        if isinstance(coord_x, int):
            self._x = coord_x
        else:
            raise ValueError('This is not an integer number. Please enter an integer: ')
    
    @property
    def y(self):
        return self._y
    
    @y.setter
    def set_y(self, coord_y):
        # Evaluate if the coord_y is an int
        if isinstance(coord_y, int):
            self._y = coord_y
        else:
            raise ValueError('This is not an integer number. Please enter an integer: ')
        
    def distance(self, coord):
        if isinstance(coord, Coordinate):
            res_x = (self._x + coord._x) ** 2
            res_y = (self._y + coord._y) ** 2
            resul = (res_x + res_y) ** 0.5
            return resul
        raise ValueError('This is not a valid coordinate')
    
    def __str__(self):
        return f'x: {self._x} - y: {self._y}'

if __name__ == '__main__':
    coord_one = Coordinate()
    coord_one.set_x = 4
    coord_one.set_y = 5

    coord_two = Coordinate()
    coord_two.set_x = -2
    coord_two.set_y = -4

    distance_one_two = round(coord_one.distance(coord_two), 2)
    print(f'The distance between coordinate one and coordinate two is {distance_one_two}')

    distance_two_one = round(coord_two.distance(coord_one), 2)
    print(f'The distance between coordinate two and coordinate one is {distance_two_one}')
```

## Herencia

- Permite crear jerarquias de clases
- Permite compartir caracteristicas entre clases 

```
class Rectangle:

    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
    
    def area(self):
        return self.base * self.altura

class Square(Rectangle):

    def __init__(self, lado):
        super().__init__(lado, lado)

if __name__ == '__main__':
    rectangle = Rectangle(3, 4)
    print(rectangle.area())

    cuadrado = Square(lado=5)
    print(cuadrado.area())
```

## Polimorfismo 

Por definición significa multiples formas. En Python es la posibilidad de cambiar el comportamiento de una superclase para adaptarlo a la subclase.

```
class Persona:

    def __init__(self, nombre):
        self.nombre = nombre 
    
    def avanzar(self):
        print(f'{self.nombre} camina')

class Ciclista(Persona):

    def __init__(self, nombre):
        super().__init__(nombre)
    
    def avanzar(self):
        print(f'{self.nombre} se mueve en su bicicleta')

def main():
    persona = Persona('David')
    persona.avanzar()

    ciclista = Ciclista('Daniel')
    ciclista.avanzar()

if __name__ == '__main__':
    main()
```

## Introducción a la complejidad algorítmica

- La complejidad algoritmica permite evaluar la eficiencia de un algoritmo. 
- La complejidad puede ser temporal o espacial
- La complejidad se puede definir como T(n)

### Aproximaciones

La forma correcta es contar los pasos conforme nos aproximamos al infinito

```
import time

def factorial(n):
    
    respuesta = 1
    
    while n > 1:
        respuesta *= n
        n -= 1
    
    return respuesta

def factorial_r(n):
    if n == 1:
        return 1
    
    return n * factorial(n-1)

if __name__ == '__main__':
    n = 1000000
    comienzo = time.time()
    factorial(n)
    final = time.time()
    print(final - comienzo)

    comienzo = time.time()
    factorial_r(n)
    final = time.time()
    print(final - comienzo)
```

##  Big O Notation

La notación de Big O permite evaluar la eficiencia de un algoritmo de acuerdo a los pasos de un algoritmo cuando este tiende a infinito.

Las clases de complejidad algoritmicas son:

* $O(1)$ es constante
* $O(n)$ es lineal
* $O(log(n))$ es logaritmico
* $O(n log(n))$ es logaritmica lineal
* $O(n^a)$ es polinomial
* $O(a^n)$ es exponencial
* $O(n!)$ es factorial
